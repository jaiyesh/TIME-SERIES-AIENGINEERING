```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Temp data.csv')
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
pd.plotting.lag_plot(df['TEMP'],lag = 1)
plt.plot([-20,50],[-20,50],c='r')
plt.title('1.')
plt.xlabel('Temperature at t')
plt.ylabel('Temperature at (t+1 hours)')
plt.grid(True)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
pd.plotting.lag_plot(df['TEMP'],lag = 8760)
plt.plot([-20,50],[-20,50],c='r')

plt.title('2.')
plt.xlabel('Temperature at t')
plt.ylabel('Temperature at (t+1 year)')
plt.grid(True)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
pd.plotting.lag_plot(df['TEMP'],lag = 4320)
plt.plot([-20,50],[50,-20],c='r')

plt.title('3.')
plt.xlabel('Temperature at t')
plt.ylabel('Temperature at (t+6 months)')
plt.grid(True)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
pd.plotting.lag_plot(df['TEMP'],lag = 2150)

plt.title('4.')
plt.xlabel('Temperature at t')
plt.ylabel('Temperature at (t+3 months)')
plt.grid(True)
```
